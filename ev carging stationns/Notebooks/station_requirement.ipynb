```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Dataset: State-wise requirement level (5 = Critical, 1 = Well-equipped)
data = {
    'State': [
        'Arunachal Pradesh', 'Nagaland', 'Meghalaya', 'Manipur', 'Mizoram',
        'Uttar Pradesh', 'Bihar', 'Rajasthan', 'West Bengal', 'Madhya Pradesh',
        'Chhattisgarh', 'Jharkhand', 'Assam', 'Odisha',
        'Delhi', 'Maharashtra', 'Karnataka', 'Gujarat', 'Tamil Nadu'
    ],
    'Requirement_Level': [
        5, 5, 5, 5, 5,
        4, 4, 3, 3, 3,
        2, 2, 2, 2,
        1, 1, 1, 1, 1
    ]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plot Heatmap
plt.figure(figsize=(10, 8))
heatmap_data = df.pivot_table(index='State', values='Requirement_Level')
sns.heatmap(
    heatmap_data,
    annot=True,
    cmap='YlOrRd',
    linewidths=0.5,
    cbar_kws={'label': 'Requirement Level'}
)

plt.title('EV Charging Station Requirement Level by State (India - 2025)')
plt.xlabel('Requirement Level')
plt.ylabel('State')
plt.tight_layout()
plt.gca().invert_yaxis()  # Critical states on top
plt.show()
```
